```python
import pdfplumber
import pandas as pd
import re

# Function to extract text from the PDF using pdfplumber
def extract_text_with_pdfplumber(file_path):
    text = ""
    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Function to extract financial information based on keywords
def extract_financial_info(text, keywords):
    extracted_info = []
    lines = text.split('\n')
    for line in lines:
        if any(keyword.lower() in line.lower() for keyword in keywords):
            extracted_info.append(line)
    return extracted_info

# Function to parse extracted financial information into a tabular format
def parse_financial_info(info_list):
    parsed_data = []
    for info in info_list:
        # Use regex to capture the metric and the value
        match = re.search(r'([a-zA-Z\s]+):\s*([\d\.\,\$B]+)', info)
        if match:
            metric = match.group(1).strip()
            value = match.group(2).strip()
            parsed_data.append({"Metric": metric, "Value": value})
    return parsed_data

# Define keywords
keywords = ['total revenue', 'total sales', 'net sales', 'net revenue', 'net income']

# Path to the uploaded PDF file
pdf_path = "Alphabet Inc..pdf"

# Extract text from the provided PDF
pdf_text = extract_text_with_pdfplumber(pdf_path)

# Extract financial information based on keywords
financial_info = extract_financial_info(pdf_text, keywords)

# Parse the extracted financial information into a tabular format
parsed_data = parse_financial_info(financial_info)

# Convert parsed data to a DataFrame
df = pd.DataFrame(parsed_data)

# Save the DataFrame to an Excel file
output_path = "financial_info.xlsx"
df.to_excel(output_path, index=False)

print(f"Financial information extracted and saved to {output_path}")
```

```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import Dataset, DataLoader

# Define a simple dataset (this is just a placeholder, actual dataset should be prepared)
class FinancialDataset(Dataset):
    def __init__(self, texts, labels):
        self.texts = texts
        self.labels = labels

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        return {'text': self.texts[idx], 'label': self.labels[idx]}

# Example data (replace with actual data preparation)
texts = [
    "Revenue for the year ended December 31, 2023 was $1,000,000.",
    "In 2023, our sales reached $2,000,000.",
]
labels = [1, 1]  # 1 for relevant, 0 for not relevant

dataset = FinancialDataset(texts, labels)
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)

# Load pre-trained BERT model
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Fine-tune BERT
optimizer = AdamW(model.parameters(), lr=1e-5)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model.to(device)
model.train()

num_epochs = 3
for epoch in range(num_epochs):
    for batch in dataloader:
        inputs = tokenizer(batch['text'], padding=True, truncation=True, return_tensors="pt")
        inputs.to(device)
        labels = batch['label'].to(device)

        outputs = model(**inputs, labels=labels)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

# Save the fine-tuned model
model.save_pretrained('./fine_tuned_model')
```
