```python
import os
import streamlit as st
import tempfile
import re
import camelot
import pandas as pd
import pdfplumber
from pikepdf import Pdf, new
def save_uploaded_file(uploaded_file, dir_name="tempDir"):
    os.makedirs(dir_name, exist_ok=True)
    path = os.path.join(dir_name, uploaded_file.name)
    with open(path, "wb") as f:
        f.write(uploaded_file.getbuffer())
    return path
def extract_pages_and_tables(pdf_path, primary_keywords, secondary_keywords):
    relevant_pages = []
    secondary_keyword_pages = []
    with pdfplumber.open(pdf_path) as pdf:
        for num, page in enumerate(pdf.pages, start=1):
            text = page.extract_text()
            if text and any(kw.lower() in text.lower() for kw in primary_keywords):
                relevant_pages.append(num)
                if any(kw.lower() in text.lower() for kw in secondary_keywords):
                    secondary_keyword_pages.append(num)
    if not relevant_pages:
        return None, None
    tables_extracted, temp_pdf_path = extract_tables_from_pdf(pdf_path, secondary_keyword_pages, secondary_keywords)
    return tables_extracted, temp_pdf_path
def extract_tables_from_pdf(pdf_path, page_numbers, secondary_keywords):
    with Pdf.open(pdf_path) as pdf:
        output_pdf = new()
        for page_num in page_numbers:
            output_pdf.pages.append(pdf.pages[page_num - 1])
        temp_pdf_path = tempfile.mktemp(suffix=".pdf", dir="tempDir")
        output_pdf.save(temp_pdf_path)
        output_pdf.close()
    tables = camelot.read_pdf(temp_pdf_path, pages='all', flavor='stream')
    tables_extracted = []
    for table in tables:
        df = table.df
        df = df.replace(['\r', '\n', '\t'], ' ', regex=True)
        # Step 2: Get units and save them
        unit_info = check_and_save_units(df)
        st.write("Units Information:")
        st.write(unit_info)
        # Step 3: Join rows which are Eigenmittel or split rows
        df = join_eigenmittel_rows(df)
        # Step 4: Join first 6 rows for column names
        new_columns = [' '.join(df.iloc[:6, col].astype(str).values) for col in range(df.shape[1])]
        df.columns = new_columns
        df = df[6:].reset_index(drop=True)
        df.columns = make_columns_unique(df.columns)
        # Convert numeric columns to numeric types
        df = df.apply(pd.to_numeric, errors='ignore')
        tables_extracted.append((page_numbers[table.page - 1], df, unit_info))
    return tables_extracted, temp_pdf_path
def check_and_save_units(df):
    unit_info = {}
    for col in range(df.shape[1]):
        if df.iloc[:, col].astype(str).str.contains('Tsd €').any():
            unit_info[col] = 'Tsd €'
        elif df.iloc[:, col].astype(str).str.contains('Millionen €').any():
            unit_info[col] = 'Millionen €'
    return unit_info
def join_eigenmittel_rows(df):
    result = []
    skip_next = False
    for i in range(len(df)):
        if skip_next:
            skip_next = False
            continue
        # Check if the row contains 'Für die Erfüllung' and next row contains 'anrechnungsfähige Eigenmittel'
        if 'Für die Erfüllung' in df.iloc[i, 0] and i < len(df) - 1 and 'anrechnungsfähige Eigenmittel' in df.iloc[i + 1, 0]:
            combined_row = df.iloc[i] + ' ' + df.iloc[i + 1]
            result.append(combined_row)
            skip_next = True
        elif 'Eigenmittel' in df.iloc[i, 0]:
            if i > 0:
                result[-1] = result[-1] + ' ' + df.iloc[i]
            else:
                result.append(df.iloc[i])
        else:
            result.append(df.iloc[i])
    return pd.DataFrame(result)
def make_columns_unique(columns):
    unique_columns = []
    counts = {}
    for col in columns:
        if col in counts:
            counts[col] += 1
            unique_columns.append(f"{col}_{counts[col]}")
        else:
            counts[col] = 0
            unique_columns.append(col)
    return unique_columns
def multiply_by_thousand_if_needed(df, unit_info, keywords):
    for keyword in keywords:
        keyword_columns = [col for col in df.columns if keyword in col]
        for col in keyword_columns:
            col_index = df.columns.get_loc(col)
            if col_index in unit_info and unit_info[col_index] == 'Tsd €':
                df[col] = df[col].apply(pd.to_numeric, errors='coerce') * 1000
    return df
def display_subkeyword_context(tables_found, secondary_keywords, pdf_name):
    combined_df = pd.DataFrame()
    for page_number, df, unit_info in tables_found:
        st.write(f"Main DataFrame from original page {page_number} in {pdf_name}:")
        st.dataframe(df)
        for sub_keyword in secondary_keywords:
            if df.astype(str).apply(lambda x: x.str.contains(sub_keyword, regex=True, na=False)).any().any():
                df_filtered = df[df.astype(str).apply(lambda row: row.str.contains(sub_keyword, case=False, na=False).any(), axis=1)]
                df_filtered = multiply_by_thousand_if_needed(df_filtered, unit_info, secondary_keywords)
                # Convert unit_info to a string
                unit_info_str = ', '.join([f"{col}: {unit}" for col, unit in unit_info.items()])
                # Using .loc to avoid SettingWithCopyWarning
                df_filtered.loc[:, 'units'] = unit_info_str
                combined_df = pd.concat([combined_df, df_filtered], ignore_index=True)
    # Remove duplicate rows
    combined_df = combined_df.drop_duplicates()
    # Display the combined DataFrame
    if not combined_df.empty:
        st.write(f"Combined DataFrame for all subkeywords in {pdf_name}:")
        st.dataframe(combined_df)
        # Save the combined DataFrame to a CSV file with the PDF name
        csv_file_name = f"{os.path.splitext(pdf_name)[0]}_combined.csv"
        combined_df.to_csv(csv_file_name, index=False)
        st.write(f"Combined DataFrame saved as {csv_file_name}")
    else:
        st.write(f"No matching subkeywords found in the extracted tables from {pdf_name}.")
# Streamlit app
st.set_page_config(layout="wide")
st.title("PDF Keyword and Table Extraction")
uploaded_files = st.file_uploader("Upload PDF files", type=["pdf"], accept_multiple_files=True)
# Define keywords and subkeywords within the code
primary_keywords = ["Betrag mit"]
secondary_keywords = ["Basiseigenmittel","SCR"]
if uploaded_files:
    if st.button("Extract Data from Uploaded Files"):
        for uploaded_file in uploaded_files:
            pdf_path = save_uploaded_file(uploaded_file)
            tables_extracted, temp_pdf_path = extract_pages_and_tables(pdf_path, primary_keywords, secondary_keywords)
            if tables_extracted:
                st.write(f"Tables containing secondary keywords in {uploaded_file.name}:")
                display_subkeyword_context(tables_extracted, secondary_keywords, uploaded_file.name)
            else:
                st.write(f"No relevant pages or tables found in the uploaded PDF {uploaded_file.name}.")
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
import google.generativeai as genai
import re
from collections import defaultdict
from google.colab import files

# Set up API key
Gemini_API_key = "AIzaSyAknSAx-pQ4n8ClUacUWaYWw-VroyXkxIA"

# Initialize Sentence Transformer model
sentence_model = SentenceTransformer('all-MiniLM-L6-v2')

# Load and process PDF
def load_and_process_pdfs(pdf_file):
    with open("temp.pdf", "wb") as f:
        f.write(pdf_file.read())
    loader = PyPDFLoader("temp.pdf")
    docs = loader.load()
    return docs

# Prepare vector store
def prepare_vectorstore(docs):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=500)
    chunks = text_splitter.split_documents(docs)
    embedding_function = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2", model_kwargs={'device': 'cpu'})
    vectorstore = Chroma.from_documents(chunks, embedding_function, persist_directory="./chroma_db_nccn")
    return vectorstore

# Get relevant context from vector store
def get_relevant_context_from_db(vectorstore):
    context = ""
    search_results = vectorstore.similarity_search("total revenue", k=6)
    for result in search_results:
        context += result.page_content + "\n"
    return context

# Generate RAG prompt
def generate_rag_prompt(context):
    escaped = context.replace("'", "").replace('"', "").replace("\n", " ")
    prompt = f"""
    Extract the most recent total revenue information from the uploaded PDF, focusing on revenue by segment or category.
    Provide the exact revenue figure for the latest year mentioned.
    Context: '{escaped}'
    """
    return prompt

# Generate answer using generative AI
def generate_answer(prompt):
    genai.configure(api_key=Gemini_API_key)
    model = genai.GenerativeModel(model_name='gemini-pro')
    answer = model.generate_content(prompt)
    return answer.text

# Extract and filter revenue data
def extract_and_filter_revenue(text, keywords):
    extracted_data = defaultdict(set)
    lines = text.split("\n")
    for line in lines:
        if any(keyword.lower() in line.lower() for keyword in keywords):
            number_pattern = re.compile(r'\b\d[\d,.]*\b', re.IGNORECASE)
            numbers = number_pattern.findall(line)
            for num in numbers:
                clean_number = num.replace(',', '')
                try:
                    value = float(clean_number)
                    if len(str(int(value))) > 4:
                        extracted_data['revenue'].add(value)
                except ValueError:
                    continue
    return extracted_data

# Main function to run in Colab
def main():
    print("Upload PDF files:")
    uploaded_files = files.upload()

    final_data = defaultdict(list)

    for file_name, file_data in uploaded_files.items():
        print(f"Processing {file_name}...")
        docs = load_and_process_pdfs(file_data)
        vectorstore = prepare_vectorstore(docs)

        context = get_relevant_context_from_db(vectorstore)
        if context:
            prompt = generate_rag_prompt(context)
            ai_answer = generate_answer(prompt)

            # Filtering the extracted information to find the most relevant revenue numbers
            keywords = ["Total net sales", "Total net revenue", "Total revenues"]
            filtered_data = extract_and_filter_revenue(ai_answer, keywords)

            if filtered_data['revenue']:
                final_data['Company'].append(file_name.replace('.pdf', ''))
                final_data['Revenue'].append(max(filtered_data['revenue']))
            else:
                final_data['Company'].append(file_name.replace('.pdf', ''))
                final_data['Revenue'].append(None)

    # Create a DataFrame and display it
    df = pd.DataFrame(final_data)
    print("Extraction complete. Here is the extracted data:")
    print(df)

    # Save the DataFrame to an Excel file
    output_excel_path = "revenue_extracted.xlsx"
    df.to_excel(output_excel_path, index=False)

    # Provide download link
    files.download(output_excel_path)

# Run the app
if __name__ == "__main__":
    main()
```

```python
import PyPDF2
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os
from langchain_google_genai import GoogleGenerativeAIEmbeddings
import google.generativeai as genai
import chromadb
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains.question_answering import load_qa_chain
from langchain.prompts import PromptTemplate
import pandas as pd
from google.colab import files
import io

# Hard-coded API key (replace with your actual API key)
GOOGLE_API_KEY = "your_google_api_key_here"
genai.configure(api_key=GOOGLE_API_KEY)

def get_pdf_text(pdf_docs):
    text = ""
    for pdf in pdf_docs:
        pdf_reader = PdfReader(pdf)
        for page in pdf_reader.pages:
            text += page.extract_text()
    return text

def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
    chunks = text_splitter.split_text(text)
    return chunks

def get_vector_store(text_chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", api_key=GOOGLE_API_KEY)

    # Initialize Chroma
    client = chromadb.Client()
    collection = client.create_collection("pdf_revenue")

    # Add vectors to Chroma
    vectors = embeddings.embed_documents(text_chunks)
    for i, chunk in enumerate(text_chunks):
        collection.add(vectors[i], metadata={"text": chunk})

def get_conversational_chain():
    prompt_template = """
    Extract the total revenue and segmented revenue for the company mentioned in the context.
    Provide the results in the format: 'Company Name: [company name], Segment: [segment], Revenue: [revenue]'.
    If the information is not available, say 'Information not available'.

    Context:\n {context}?\n
    """

    model = ChatGoogleGenerativeAI(model="gemini-pro", temperature=0.3)
    prompt = PromptTemplate(template=prompt_template, input_variables=["context", "question"])
    chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)

    return chain

def extract_revenue_data(query):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
    client = chromadb.Client()
    collection = client.get_collection("pdf_revenue")

    # Query Chroma
    query_vector = embeddings.embed_query(query)
    results = collection.query(query_vector, top_k=5)  # Adjust top_k as needed

    chain = get_conversational_chain()
    extracted_results = []

    for result in results['documents']:
        response = chain({"input_documents": [result['metadata']['text']], "question": query}, return_only_outputs=True)
        extracted_results.append(response["output_text"])

    return extracted_results

def save_to_excel(data, output_path):
    df = pd.DataFrame(data, columns=["Company Name & Segment", "Revenue"])
    df.to_excel(output_path, index=False)
    return output_path

def process_pdfs(pdf_docs):
    raw_text = get_pdf_text(pdf_docs)
    text_chunks = get_text_chunks(raw_text)
    get_vector_store(text_chunks)

    query = "What is the total revenue and segmented revenue?"
    revenue_data = extract_revenue_data(query)

    structured_data = []

    for entry in revenue_data:
        for line in entry.split('\n'):
            if "Company Name" in line:
                company_info, revenue = line.split(', Revenue: ')
                structured_data.append([company_info.strip(), revenue.strip()])

    excel_path = save_to_excel(structured_data, "company_revenue.xlsx")
    return excel_path

# Upload PDFs
uploaded_files = files.upload()
pdf_docs = [io.BytesIO(uploaded_file) for uploaded_file in uploaded_files.values()]

# Process PDFs and save results
excel_file_path = process_pdfs(pdf_docs)
files.download(excel_file_path)
```

```python
import pandas as pd
from google.colab import files
import google.generativeai as genai
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from PyPDF2 import PdfReader
import io

# API key for Google Generative AI
api_key = "YOUR_GOOGLE_API_KEY"  # Replace with your actual API key

# Configure Google Generative AI
genai.configure(api_key=api_key)

def upload_files():
    """
    Upload Excel and Taxonomy document files.
    """
    print("Please upload the Excel sheet:")
    excel_file = files.upload()
    print("Please upload the Taxonomy document:")
    taxonomy_file = files.upload()

    excel_path = next(iter(excel_file))
    taxonomy_path = next(iter(taxonomy_file))

    return excel_path, taxonomy_path

def extract_text_from_pdf(pdf_path):
    """
    Extract text from a PDF file.
    """
    pdf_text = ""
    with open(pdf_path, "rb") as file:
        pdf_reader = PdfReader(file)
        for page in pdf_reader.pages:
            pdf_text += page.extract_text()
    return pdf_text

def extract_taxonomy(taxonomy_path):
    """
    Extract activities from the taxonomy PDF document.
    """
    taxonomy_text = extract_text_from_pdf(taxonomy_path)
    # Split the extracted text into lines (assuming each line is an activity)
    taxonomy = taxonomy_text.splitlines()
    return taxonomy

def check_description_eligibility(description, taxonomy):
    """
    Check if the description matches any activity in the taxonomy using a prompt-based approach.
    """
    # Create a prompt template
    prompt_template = """
    Given the following description:
    {description}

    and the list of activities:
    {taxonomy}

    Determine if the description closely matches any of the activities based on meaning and context. Return "Yes" if there is a close match, otherwise return "No".
    """

    prompt = prompt_template.format(description=description, taxonomy="\n".join(taxonomy))

    # Initialize the model
    model = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0.3, google_api_key=api_key)

    # Query the model
    response = model(prompt)

    # Extract and return the response
    result = response['response'].strip()
    return result

def update_excel_with_eligibility(excel_path, taxonomy):
    """
    Update the Excel sheet with an Eligibility column.
    """
    # Load the Excel file
    df = pd.read_excel(excel_path)

    # Check if the required columns exist
    if 'Description' not in df.columns:
        raise ValueError("The Excel sheet must contain a 'Description' column.")

    # Determine eligibility for each description
    df['Eligibility'] = df['Description'].apply(lambda desc: check_description_eligibility(desc, taxonomy))

    # Save the updated DataFrame to a new Excel file
    updated_excel_path = 'updated_revenue_data.xlsx'
    df.to_excel(updated_excel_path, index=False)

    # Download the updated Excel file
    files.download(updated_excel_path)
    print(f"Updated Excel file has been saved and downloaded as {updated_excel_path}")

def main():
    excel_path, taxonomy_path = upload_files()
    taxonomy = extract_taxonomy(taxonomy_path)
    update_excel_with_eligibility(excel_path, taxonomy)

if __name__ == "__main__":
    main()

```

```python
import io
from PyPDF2 import PdfReader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import FAISS
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.chains.question_answering import load_qa_chain
from langchain.prompts import PromptTemplate
from google.colab import files
import google.generativeai as genai
import pandas as pd

# API key
api_key = "AIzaSyBhaSId2Ddw_kn4iCaM-31_TW1Or0GjNsE"

# Configure the Google Generative AI with the API key
genai.configure(api_key=api_key)

def get_pdf_text(pdf_files):
    pdf_texts = []
    for pdf in pdf_files:
        pdf_reader = PdfReader(io.BytesIO(pdf.read()))
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text()
        pdf_texts.append(text)
    return pdf_texts

def get_text_chunks(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=10000, chunk_overlap=1000)
    chunks = text_splitter.split_text(text)
    return chunks

def get_vector_store(text_chunks):
    embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", google_api_key=api_key)
    vector_store = FAISS.from_texts(text_chunks, embedding=embeddings)
    vector_store.save_local("faiss_index")
    return vector_store

def get_conversational_chain():
    prompt_template = """
    Extract the total revenue for each company from the provided text. Additionally, identify and extract the segmented revenues contributing to the total revenue. For each segment, provide a detailed description explaining what the segment includes or represents. The description should clarify the nature of the products, services, or activities involved in the segment.

    Ensure that you include the appropriate unit (e.g., USD, $) and scale (e.g., thousands, millions, billions) with the revenue numbers. For instance, if the revenue is $22.6 million, it should be presented as "$22.6 million USD" or "$22.6 million".

    Compare each description with the criteria mentioned in the Taxonomy Document and if both have similarity, then the eligibility should be "Yes". Otherwise, it should be "No".

    Your output should be in a consistent tabular format, even if the input formats differ. The table should have the following structure:

    | Company Name | Segment Name         | Revenue             | Description                                                   |Eligibility
    |--------------|----------------------|---------------------|---------------------------------------------------------------|-------------------------------------------
    | Company A    | Total Revenue        | $X million USD      | Overall total revenue for Company A                           |
    | Company A    | Segment 1            | $Y million USD      | Detailed description of Segment 1 (e.g., "Includes software sales and support services") | Yes/No
    | Company A    | Segment 2            | $Z billion USD      | Detailed description of Segment 2 (e.g., "Comprises manufacturing of electronics and accessories") |Yes/No
    | ...          | ...                  | ...                 | ...                                                           |

    Ensure that:
    - "Total Revenue" is listed as a row under each company.
    - Each segment name is indented properly under its respective company.
    - Include the currency unit (e.g., USD, $) and scale (e.g., millions, billions) along with the revenue numbers.
    - Provide specific descriptions for each segment, explaining the core activities, products, or services.
    - Avoid generic phrases like "revenue from segment activity" and aim to provide more detailed context based on the text.

    {context}
    """
    model = ChatGoogleGenerativeAI(model="gemini-1.5-pro", temperature=0.3, google_api_key=api_key)
    prompt = PromptTemplate(template=prompt_template, input_variables=["context"])
    chain = load_qa_chain(model, chain_type="stuff", prompt=prompt)
    return chain
def main():
    # Upload PDF files
    uploaded_files = files.upload()
    pdf_files = [io.BytesIO(uploaded_files[file_name]) for file_name in uploaded_files]

    all_responses = []  # To store responses for each PDF

    if pdf_files:
        # Extract and process text from each PDF separately
        pdf_texts = get_pdf_text(pdf_files)

        # Process each PDF text separately
        for pdf_text in pdf_texts:
            text_chunks = get_text_chunks(pdf_text)
            vector_store = get_vector_store(text_chunks)

            # Load the vector store
            embeddings = GoogleGenerativeAIEmbeddings(model="models/embedding-001", google_api_key=api_key)
            new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
            docs = new_db.similarity_search("revenue")

            # Use conversational chain for QA
            chain = get_conversational_chain()
            response = chain({"input_documents": docs, "context": pdf_text}, return_only_outputs=True)

            all_responses.append(response["output_text"])

        # Display the results for all PDFs
        for idx, res in enumerate(all_responses):
            print(f"Reply for PDF {idx+1}: \n", res)

        # Export results to Excel
        if all_responses:
            results = []
            for res in all_responses:
                # Process each response (convert each line of the table into a list)
                rows = res.strip().split('\n')[1:]  # Skip the header row
                for row in rows:
                    columns = row.split('|')[1:-1]  # Extract columns between separators
                    results.append([col.strip() for col in columns])  # Remove leading/trailing whitespace

            # Create a DataFrame from the list of results
            df = pd.DataFrame(results, columns=["Company Name", "Segment Name", "Revenue", "Description"])

            # Save the DataFrame to an Excel file
            df.to_excel('revenue_data.xlsx', index=False)

            # Download the Excel file in Google Colab
            files.download('revenue_data.xlsx')

    else:
        print("No files uploaded.")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
from PyPDF2 import PdfReader
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    pdf_text = ""
    with open(pdf_path, 'rb') as file:
        reader = PdfReader(file)
        for page in reader.pages:
            pdf_text += page.extract_text()
    return pdf_text

# Load the Excel sheet and Taxonomy Document
excel_file = 'Xylem.xlsx'
taxonomy_pdf = 'Taxonomy Document.pdf'

# Load Excel sheet
df = pd.read_excel(excel_file)

# Extract text from the Taxonomy PDF
taxonomy_text = extract_text_from_pdf(taxonomy_pdf)

# Define the Prompt Template
prompt_template = """
Compare the following description from the Excel sheet with the activities listed in the taxonomy document. Determine if there is a semantic similarity between the description and any activity in the taxonomy. If the description matches an activity and the revenue is present, then the eligibility should be "Yes". Otherwise, it should be "No".

**Description:** {description}

**Taxonomy Document:**
{taxonomy_text}

**Revenue:** {revenue}

**Your Task:**
1. Check if the description semantically matches any of the activities in the taxonomy document.
2. If a match is found and revenue is provided, return "Yes" for eligibility.
3. If no match is found or no revenue is provided, return "No" for eligibility.

**Output Format:**
Eligibility: [Yes/No]
"""

# Configure the AI model
api_key = 'AIzaSyDQaIjl2BABGdx9lW0P1FZA4Txm4CsxEnE'
model = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0.3, google_api_key=api_key)
prompt = PromptTemplate(template=prompt_template, input_variables=["description", "taxonomy_text", "revenue"])

# Process each description and determine eligibility
results = []
for index, row in df.iterrows():
    description = row['Description']
    revenue = row['Revenue']

    # Generate the prompt
    full_prompt = prompt.format(description=description, taxonomy_text=taxonomy_text, revenue=revenue)

    # Get the response from the model
    response = model([{"role": "user", "content": full_prompt}])
    eligibility = response['choices'][0]['message']['content'].strip()

    # Append the result
    results.append(eligibility)

# Add results to the DataFrame
df['Eligibility'] = results

# Save the updated DataFrame to a new Excel file
df.to_excel('updated_revenue_data.xlsx', index=False)

print("Processing complete. The updated file has been saved as 'updated_revenue_data.xlsx'.")


```
